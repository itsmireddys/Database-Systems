```python
import mysql.connector
from mysql.connector import Error

def connect_fetch_patients_data():
    """Connect to the MySQL database and fetch data from the patients table."""
    try:
        # Connection to the database (update these parameters as needed)
        conn = mysql.connector.connect(
            host='localhost',
            database='hospital_data_management_system',
            user='root',
            password='Root@123'
        )

        if conn.is_connected():
            print('Successfully connected to the database')
            print('')
            cursor = conn.cursor()

            # Querying the patients table for the first 10 entries
            cursor.execute('SELECT * FROM patients LIMIT 10;')

            # Fetching the data
            patients = cursor.fetchall()
            print('patient data retrieved from the database:')
            print('')
            
            # Presenting the data on the screen
            for patient in patients:
                print(patient)

            # Storing the data to a text file
            with open('patient_data.txt', 'w') as f:
                for patient in patients:
                    f.write(str(patient) + '\n')

            print('patient data has been written to patient_data.txt')
            print('')

    except Error as e:
        print('Error:', e)

if __name__ == '__main__':
    connect_fetch_patients_data()
    
    
def connect_fetch_doctors_data():
    """Connect to the MySQL database and fetch data from the doctors table."""
    try:
        # Connection to the database (update these parameters as needed)
        conn = mysql.connector.connect(
            host='localhost',
            database='hospital_data_management_system',
            user='root',
            password='Root@123'
        )

        if conn.is_connected():
            cursor = conn.cursor()

            # Querying the doctors table for the first 10 entries
            cursor.execute('SELECT * FROM doctors LIMIT 10;')

            # Fetching the data
            doctors = cursor.fetchall()
            print('doctor data retrieved from the database:')
            print('')
            
            # Presenting the data on the screen
            for doctor in doctors:
                print(doctor)

            # Storing the data to a text file
            with open('doctor_data.txt', 'w') as f:
                for doctor in doctors:
                    f.write(str(doctor) + '\n')

            print('doctor data has been written to doctor_data.txt')
            print('')

    except Error as e:
        print('Error:', e)

if __name__ == '__main__':
    connect_fetch_doctors_data()

    
def connect_fetch_inpatients_data():
    """Connect to the MySQL database and fetch data from the inpatients table."""
    try:
        # Connection to the database (update these parameters as needed)
        conn = mysql.connector.connect(
            host='localhost',
            database='hospital_data_management_system',
            user='root',
            password='Root@123'
        )

        if conn.is_connected():
            cursor = conn.cursor()

            # Querying the inpatients table for the first 10 entries
            cursor.execute('SELECT * FROM inpatient LIMIT 10;')

            # Fetching the data
            inpatients = cursor.fetchall()
            print('inpatient data retrieved from the database:')
            print('')
            
            # Presenting the data on the screen
            for inpatient in inpatients:
                print(inpatient)

            # Storing the data to a text file
            with open('inpatient_data.txt', 'w') as f:
                for inpatient in inpatients:
                    f.write(str(inpatient) + '\n')

            print('inpatient data has been written to inpatient_data.txt')
            print('')

    except Error as e:
        print('Error:', e)

if __name__ == '__main__':
    connect_fetch_inpatients_data()

        
def connect_fetch_rooms_data():
    """Connect to the MySQL database and fetch data from the rooms table."""
    try:
        # Connection to the database (update these parameters as needed)
        conn = mysql.connector.connect(
            host='localhost',
            database='hospital_data_management_system',
            user='root',
            password='Root@123'
        )

        if conn.is_connected():
            cursor = conn.cursor()

            # Querying the rooms table for the first 10 entries
            cursor.execute('SELECT * FROM rooms LIMIT 10;')

            # Fetching the data
            rooms = cursor.fetchall()
            print('room data retrieved from the database:')
            print('')
            
            # Presenting the data on the screen
            for room in rooms:
                print(room)

            # Storing the data to a text file
            with open('room_data.txt', 'w') as f:
                for room in rooms:
                    f.write(str(room) + '\n')

            print('room data has been written to room_data.txt')
            print('')

    except Error as e:
        print('Error:', e)

if __name__ == '__main__':
    connect_fetch_rooms_data()
    
    
def connect_fetch_labs_data():
    """Connect to the MySQL database and fetch data from the labs table."""
    try:
        # Connection to the database (update these parameters as needed)
        conn = mysql.connector.connect(
            host='localhost',
            database='hospital_data_management_system',
            user='root',
            password='Root@123'
        )

        if conn.is_connected():
            cursor = conn.cursor()

            # Querying the labs table for the first 10 entries
            cursor.execute('SELECT * FROM labs LIMIT 10;')

            # Fetching the data
            labs = cursor.fetchall()
            print('lab data retrieved from the database:')
            print('')
            
            # Presenting the data on the screen
            for lab in labs:
                print(lab)

            # Storing the data to a text file
            with open('lab_data.txt', 'w') as f:
                for lab in labs:
                    f.write(str(lab) + '\n')

            print('lab data has been written to lab_data.txt')
            print('')

    except Error as e:
        print('Error:', e)

if __name__ == '__main__':
    connect_fetch_labs_data()
    
    
def connect_fetch_labtests_data():
    """Connect to the MySQL database and fetch data from the labtests table."""
    try:
        # Connection to the database (update these parameters as needed)
        conn = mysql.connector.connect(
            host='localhost',
            database='hospital_data_management_system',
            user='root',
            password='Root@123'
        )

        if conn.is_connected():
            cursor = conn.cursor()

            # Querying the labtests table for the first 10 entries
            cursor.execute('SELECT * FROM labtests LIMIT 10;')

            # Fetching the data
            labtests = cursor.fetchall()
            print('labtest data retrieved from the database:')
            print('')
            
            # Presenting the data on the screen
            for labtest in labtests:
                print(labtest)

            # Storing the data to a text file
            with open('labtest_data.txt', 'w') as f:
                for labtest in labtests:
                    f.write(str(labtest) + '\n')

            print('labtest data has been written to labtest_data.txt')
            print('')

    except Error as e:
        print('Error:', e)

if __name__ == '__main__':
    connect_fetch_labtests_data()
    
    
def connect_fetch_tests_data():
    """Connect to the MySQL database and fetch data from the tests table."""
    try:
        # Connection to the database (update these parameters as needed)
        conn = mysql.connector.connect(
            host='localhost',
            database='hospital_data_management_system',
            user='root',
            password='Root@123'
        )

        if conn.is_connected():
            cursor = conn.cursor()

            # Querying the tests table for the first 10 entries
            cursor.execute('SELECT * FROM tests LIMIT 10;')

            # Fetching the data
            tests = cursor.fetchall()
            print('test data retrieved from the database:')
            print('')
            
            # Presenting the data on the screen
            for test in tests:
                print(test)

            # Storing the data to a text file
            with open('test_data.txt', 'w') as f:
                for test in tests:
                    f.write(str(test) + '\n')

            print('test data has been written to test_data.txt')
            print('')

    except Error as e:
        print('Error:', e)

if __name__ == '__main__':
    connect_fetch_tests_data()
    
    
def connect_fetch_medications_data():
    """Connect to the MySQL database and fetch data from the medications table."""
    try:
        # Connection to the database (update these parameters as needed)
        conn = mysql.connector.connect(
            host='localhost',
            database='hospital_data_management_system',
            user='root',
            password='Root@123'
        )

        if conn.is_connected():
            cursor = conn.cursor()

            # Querying the medications table for the first 10 entries
            cursor.execute('SELECT * FROM medication LIMIT 10;')

            # Fetching the data
            medications = cursor.fetchall()
            print('medication data retrieved from the database:')
            print('')
            
            # Presenting the data on the screen
            for medication in medications:
                print(medication)

            # Storing the data to a text file
            with open('medication_data.txt', 'w') as f:
                for medication in medications:
                    f.write(str(medication) + '\n')

            print('medication data has been written to medication_data.txt')
            print('')

    except Error as e:
        print('Error:', e)

if __name__ == '__main__':
    connect_fetch_medications_data()
    
    
def connect_fetch_bills_data():
    """Connect to the MySQL database and fetch data from the bills table."""
    try:
        # Connection to the database (update these parameters as needed)
        conn = mysql.connector.connect(
            host='localhost',
            database='hospital_data_management_system',
            user='root',
            password='Root@123'
        )

        if conn.is_connected():
            cursor = conn.cursor()

            # Querying the bills table for the first 10 entries
            cursor.execute('SELECT * FROM bill LIMIT 10;')

            # Fetching the data
            bills = cursor.fetchall()
            print('bill data retrieved from the database:')
            print('')
            
            # Presenting the data on the screen
            for bill in bills:
                print(bill)

            # Storing the data to a text file
            with open('bill_data.txt', 'w') as f:
                for bill in bills:
                    f.write(str(bill) + '\n')

            print('bill data has been written to bill_data.txt')
            print('')

    except Error as e:
        print('Error:', e)

if __name__ == '__main__':
    connect_fetch_bills_data()
    
    
def connect_fetch_outpatients_data():
    """Connect to the MySQL database and fetch data from the outpatients table."""
    try:
        # Connection to the database (update these parameters as needed)
        conn = mysql.connector.connect(
            host='localhost',
            database='hospital_data_management_system',
            user='root',
            password='Root@123'
        )

        if conn.is_connected():
            cursor = conn.cursor()

            # Querying the outpatients table for the first 10 entries
            cursor.execute('SELECT * FROM outpatient LIMIT 10;')

            # Fetching the data
            outpatients = cursor.fetchall()
            print('outpatient data retrieved from the database:')
            print('')
            
            # Presenting the data on the screen
            for outpatient in outpatients:
                print(outpatient)

            # Storing the data to a text file
            with open('outpatient_data.txt', 'w') as f:
                for outpatient in outpatients:
                    f.write(str(outpatient) + '\n')

            print('outpatient data has been written to outpatient_data.txt')
            print('')

    except Error as e:
        print('Error:', e)
    finally:
        if conn is not None and conn.is_connected():
            conn.close()
            print('Database connection is closed.')
            print('')

if __name__ == '__main__':
    connect_fetch_outpatients_data()
```
